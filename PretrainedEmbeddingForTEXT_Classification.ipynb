# Let's Keep it Simple

# Using pretrained embedding from tensorflow_hub to handle imdb_review dataset


```python
# importing usefull libraries
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import pandas as pd
import tensorflow_hub as hub
```

# loading the embedding layer

```python
model = tf.keras.Sequential([
                             hub.KerasLayer("https://tfhub.dev/google/tf2-preview/nnlm-en-dim50/1",
                                            dtype=tf.string, input_shape=[], output_shape=[50]), # in this case this layer encodes each string as a 50-dimenion vector
                             tf.keras.layers.Dense(128, activation='relu'),
                             tf.keras.layers.Dense(1, activation='sigmoid')

])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# Load the data

```python
datasets, info = tfds.load("imdb_reviews", as_supervised=True, with_info=True)
train_size = info.splits['train'].num_examples
batch_size = 32
train_set = datasets['train'].batch(batch_size).prefetch(1)
history = model.fit(train_set, epochs=5)
```
